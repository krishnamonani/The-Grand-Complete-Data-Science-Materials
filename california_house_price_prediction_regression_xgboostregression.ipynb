House Price Prediction using regresion(XGBoost regression)

importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the boston house price dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

Loading all data into pandas dataframe to make it look structured

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data)
```

```python
#first fie rows of out dataframe
house_price_dataframe.head()
```

To make it look more understandable use columns keyword

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
#here are only DATA columns shown we need to add TARGET columns too
```

```python
# add the target(price) columns to the DataFrame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checking number of rows and columns in dataframe
house_price_dataframe.shape
```

```python
house_price_dataframe.tail()
```

Checking for any missing values

```python
#check for missing vlas
house_price_dataframe.isnull()
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.head(

)
```

Statastical measures of the dataset

```python
house_price_dataframe.describe()
```

Understanfing the corelation between various features in the dataset

1.Positive correlation
2.Negative correlation

```python
correlation = house_price_dataframe.corr()
```

```python
#constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Oranges')
```

Splitting data and target

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y =house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Split into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Model training


XGBoost regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#R square error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

#Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error:",score_1)#value should be closer to 0
print("Mean absolute error:",score_2)
```

Here, out prediction values are small that means our model is performing well.

Visualizing the actual prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual price vs Predicted price")
plt.show()
```

Prediction on test data

```python
#accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```
